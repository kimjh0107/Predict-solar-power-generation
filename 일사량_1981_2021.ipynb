```python
prophet_df = df.rename({'일시_datetime':'ds', '일사(MJ/m2)':'y'}, axis='columns')
df_val_true = prophet_df[(prophet_df['ds']>='2021-02-01 01:00:00')&(prophet_df['ds']<='2021-03-1 00:00:00')]['y'].to_numpy()
prophet_df = prophet_df[prophet_df['ds']<='2021-02-01 00:00:00']
m = Prophet(seasonality_mode='additive', 
            yearly_seasonality=20, 
            weekly_seasonality=10, 
            daily_seasonality=25,
            changepoint_prior_scale=0.05) # changepoint_pr
m.fit(prophet_df)

future = m.make_future_dataframe(periods=28*24, freq='H')
forecast = m.predict(future)
df_val_pred = forecast['yhat'][-28*24:].to_numpy()
r2_score(df_val_true, df_val_pred) # 예측을 잘 했는지 확인하기 위해 r2_score 체크
```
